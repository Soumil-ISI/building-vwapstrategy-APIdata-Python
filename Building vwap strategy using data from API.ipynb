```python
import yfinance as yf
import pandas as pd 
data = yf.download('AAPL')
data 
```

```python
#This gives us the entire column titled close
data['Close']
```

```python

print(data['Close'][0])
len(data)
```

```python
close_prices = []
volume = []

for i in range(len(data)):
    cl = data['Close'][i]
    vol= data['Volume'][i]
    close_prices.append(cl)
    volume.append(vol)

    
```

```python
close_prices
```

```python
volume
```

```python
#slicing
print(volume[2:10])
print(volume[:5]) #gives from start to 4. as it starts from 0
print(volume[-5:]) #gives last 5 elements in the list
volume[10700:] #gives from 10700 to end
```

```python
import numpy as np
```

```python
#we can multiply lists elementwise using numpy. 
np.multiply(close_prices, volume)

#our output will be an array, not a list
```

```python
#vwap is sum of daily_volume*daily_price divided by total_volume
np.sum(np.multiply(close_prices, volume))/np.sum(volume)
```

```python
#vwap of the 1st 20 days 
np.sum(np.multiply(close_prices[:20], volume[:20]))/np.sum(volume[:20])
```

```python
#vwap of last 20 days. (moving vwap since api data keeps updating as days pass)
vwap_last_20 = np.sum(np.multiply(close_prices[-20:], volume[-20:]))/np.sum(volume[-20:])
vwap_last_20
```

```python
close_prices[-1]
#value of last day's close 
```

```python
if vwap_last_20 > close_prices[-1]:
    print("BUY")
else: print("SELL")
```
